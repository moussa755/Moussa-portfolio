### `Import (numpy, pandas, matplotlib, seaborn, os) by their standard aliases`

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import os 
```

### `Create two lists of : years & durations, then a Dictionary, then a DataFrame`

```python
## Create the years and durations lists
## Assign the variables names to (years, durations)

Years=[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
Durations=[103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
print(f'Years --> {Years} ')
print('***************************************************************************')
print(f'Durations --> {Durations}')
```

```python
## Create a dictionary with the two lists --> varibale name is (movie_dict)
movie_dict = {'Years': [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
'Durations':[103, 101, 99, 100, 100, 95, 95, 96, 93, 90]}
movie_dict
```

```python
## Create a DataFrame from the dictionary --> dataframe name is (df_durations)
df_durations = pd.DataFrame(movie_dict)
df_durations
```

### `Create a LinePlot from the two columns of the above DataFrame`

`use any library and any method`
`Try to understand what from the graph if something is happening -- statistical thinking`

```python
## Draw a line plot of years and durations
## Don't forget to rename axis and make a title --> make your graph is readable and interesting
plt.figure(figsize=(10,5))
sns.lineplot(x='Years', y='Durations', data=df_durations,marker='o',color='red')
plt.title('Realtion between Years and Durations')
plt.grid(True)
plt.show()
```

### `Slice the (df_netflix) DF, by getting only type=='Movie'`

* `I want you to get a DF, sliced from main DF (df_netflix), by getting it when (type='Movie')`

```python
df_netflix.columns
```

```python
## slice the (df_netflix), when (type=='Movie'),  getting a new DF named (df_movied)
## Show the head of the new DF
## Aftrer doing that, check the unique values again of the column (type), it must contains (Movie) 

df_movied=df_netflix[df_netflix['type']=='Movie']
df_movied.head(6)
```

### `Explore the new DF, it contains two columns (release_year, durations)`

### `Going Deeper, what about getting short movies`

`Slice from (df_movies) the short movies which duration is less than 60 minutes --> new DF (df_short)`

```python
## slice from (df_movies) the short movies which their duration < 60 minutes
## Assign the new DF to variable named (df_short)
## show the head of new DF (df_short)

df_short=df_movied[df_movied['duration']< 60]
df_short.head(4)
```

### `From (df_short), get only the below columns (slice the desired columns only)`

* `Get these columns only : ['title', 'country', 'genre', 'release_year', 'duration']`
* `Assign the DF name to the same name (df_short), I mean override or (inplace=True)`

```python
df_short.columns
```

```python
## getting the desired columns only ['title', 'country', 'genre', 'release_year', 'duration'] from (df_short)
## Show the Head  again after selecting the desired columns only 

df_short[['title', 'country', 'release_year','duration', 'genre' ]].head(5)
```

### `Create Scatter plot but make (hue='genre'), it is prefreable here to use (seaborn)`
* `Scatter plot between (releaes_year) & (duration) columns from (df_short)`
* `you can use the relplot in seaborn`

```python
## Scatter plot between (releaes_year) & (duration) columns from (df_short)
## Don't forget the axis labels and title 

plt.figure(figsize=(10,5))
sns.scatterplot(x='release_year', y='duration', data=df_short, hue='genre')

plt.title('Relation between Release Years and Durations for short Movies', fontsize=14)
plt.xlabel('Relation Years', fontsize=14)
plt.ylabel('Durations', fontsize=14)
plt.legend(title='genre',bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
