```python
class Employee :
    def __init__(self, id , name , age , salary):

        if age < 0:
            raise ValueError("Age cannot be negative")

        if salary < 0:
            raise ValueError("Salary cannot be negative")

        self.id = id 
        self.name = name
        self.age = age
        self._salary = salary

    def calculate_salary(self):
        return self._salary

    def display_info(self):
        print(f"ID: {self.id}")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Salary: {self.calculate_salary()}")
        print("-" * 30)
```

```python
class Manager(Employee):

    def __init__(self , id , name , age , salary, bonus , department):

        if bonus < 0:
            raise ValueError("Bonus cannot be negative")

        super().__init__(id , name , age , salary)

        self.bonus = bonus
        self.department = department

    def calculate_salary(self):
        return self._salary + self.bonus

    def display_info(self):
        print("Manager Employee")
        print(f"ID: {self.id}")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Department: {self.department}")
        print(f"Bonus: {self.bonus}")
        print(f"Final Salary: {self.calculate_salary()}")
        print("-" * 30)
```

```python
class Developer (Employee):
    def __init__(self , id , name , age , salary , programming_language , overtime_hours):
        super().__init__( id , name , age , salary)
        self.programming_language = programming_language
        self.overtime_hours = overtime_hours
```
